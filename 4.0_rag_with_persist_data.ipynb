# **RAG IMPLEMENTATION WITH Persisted Data**

#### **Step-1: Creating Embeddings for the chunks.**

```python
from langchain_openai import OpenAIEmbeddings

embedding_model = OpenAIEmbeddings(
    model="text-embedding-3-small",
    check_embedding_ctx_length=False,
)
```

#### **Step-2: Re-Use vector DB.**

```python
from langchain_community.vectorstores import Chroma

vectorstore_persisted = Chroma(persist_directory="./vectorstore", embedding_function=embedding_model)
```

#### **Step-3: Semantic Search.**

```python
vectorstore_persisted.similarity_search(query="What is the experience with Python?", k=2)
```
